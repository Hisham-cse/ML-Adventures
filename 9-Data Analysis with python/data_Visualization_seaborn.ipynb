### Data Visualization with seaborn

Seaborn is a python visualization library based on matplotlib that provides a high level interface for drawing attractive and informative graphics .seaborn helps in crating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
## Basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter plot of Total bill vs tip ")
plt.show()
```

```python
## line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of Total bill vs size ")
plt.show()
```

```python
## categorical plots
#bar plot
sns.barplot(x="day",y="total_bill",data=tips)
plt.title("Bar plot of Total bill by day ")
plt.show()
```

```python
## Box Plot
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
## Violin plot
sns.violinplot(x="day",y="total_bill",data=tips)
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## KDE plot
sns.kdeplot(tips['total_bill'],fill="True")
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
# Heatmap
corr = tips[['total_bill','tip','size']].corr()
corr
# sns.heatmap(corr)
```

```python
import pandas as pd
sales_df = pd.read_csv('data.csv')
sales_df.head()
```

```python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x="Category",y="Value",data=sales_df,estimator=sum)
plt.title("Scatter plot of Total bill vs tip ")
plt.xlabel('product')
plt.ylabel('sales')
plt.show()
```

```python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x="Region",y="Product",data=sales_df,estimator=sum)
plt.title("Scatter plot of Total bill vs tip ")
plt.xlabel('product')
plt.ylabel('sales')
plt.show()
```
